```python
import pandas as pd
import numpy as np
sus_goals=pd.read_csv("/content/drive/MyDrive/Data Analysis With python/Top20country_sus_goals/Top20country_unsusdev_data.csv")
```

```python
sus_goals.head()
```

```python
# Forward fill the missing values for columns representing the years
sus_goals.iloc[:, 4:] = sus_goals.iloc[:, 4:].ffill(axis=0)
```

```python
# Replace '..' with NaN
sus_goals.replace('..', np.nan, inplace=True)
```

```python
# Forward fill the missing values for columns representing the years
sus_goals.iloc[:, 4:] = sus_goals.iloc[:, 4:].ffill(axis=0)
```

```python
# List of invalid country names to remove
invalid_entries = [
    'Data from database: Sustainable Development Goals (SDGs)',
    'Last Updated: 07/22/2022',
    'nan',  # to handle actual NaN values
    None
]

# Clean the 'Country Name' column by removing invalid entries
sus_goals = sus_goals[~sus_goals['Country Name'].isin(invalid_entries)]

# Drop any rows with NaN in 'Country Name' to ensure no empty country names remain
sus_goals = sus_goals.dropna(subset=['Country Name'])
```

```python
# List of year columns to be converted
year_columns = [
    '1990 [YR1990]', '1991 [YR1991]', '1992 [YR1992]', '1993 [YR1993]', '1994 [YR1994]',
    '1995 [YR1995]', '1996 [YR1996]', '1997 [YR1997]', '1998 [YR1998]', '1999 [YR1999]',
    '2000 [YR2000]', '2001 [YR2001]', '2002 [YR2002]', '2003 [YR2003]', '2004 [YR2004]',
    '2005 [YR2005]', '2006 [YR2006]', '2007 [YR2007]', '2008 [YR2008]', '2009 [YR2009]',
    '2010 [YR2010]', '2011 [YR2011]', '2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]',
    '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]',
    '2020 [YR2020]'
]

# Convert each year column from object to float
for column in year_columns:
    sus_goals[column] = pd.to_numeric(sus_goals[column], errors='coerce')
```

```python
# Ensure that 'numeric_cols' contains a list of the column names
numeric_cols = sus_goals.select_dtypes(include=['float64']).columns

# Fill NaN values in numeric columns with the mean of the respective column
sus_goals[numeric_cols] = sus_goals[numeric_cols].apply(lambda col: col.fillna(col.mean()))
```
