```python
SDG = pd.melt(sus_goal,
                  id_vars=["Country Name", "Country Code", "Series Name", "Series Code"],
                  var_name="Year",
                  value_name="Value")

# Clean the 'Year' column to remove the '[YR]' part
SDG['Year'] = SDG['Year'].str.extract(r'(\d{4})').astype(int)
```

```python
SDG
```

```python
# Invert the dictionary for easier look-up: map Series Name to SDG
SDG_lookup = {series: sdg for sdg, series_list in sdg_dict.items() for series in series_list}

# Map the 'Series Name' column to SDG based on the mapping
SDG['SDG'] = SDG['Series Name'].map(SDG_lookup)

# View the dataset with the new SDG column
SDG.head()
```

```python
from google.colab import files

SDG.to_csv('SDG_Year.csv', index=False)
# Download the sus_goals.csv file
files.download('SDG_Year.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set up Seaborn style
sns.set(style="whitegrid")

# Group data by SDG
sdg_groups = SDG.groupby('SDG')

# Loop over each SDG to create a plot
for sdg, sdg_data in sdg_groups:

    # Aggregate the data for each country and year by taking the mean of all indicators in this SDG
    aggregated_data = sdg_data.groupby(['Country Name', 'Year'], as_index=False).agg({'Value': 'mean'})

    # Separate India's data from other countries
    india_data = aggregated_data[aggregated_data['Country Name'] == 'India']
    other_countries_data = aggregated_data[aggregated_data['Country Name'] != 'India']

    # Create a figure for each SDG
    plt.figure(figsize=(6, 8))

    # Plot India's aggregated time series for this SDG
    sns.lineplot(
        data=india_data,
        x='Year',
        y='Value',
        label='India',
        color='blue',
        marker='o',
        linewidth=2
    )

    # Plot the aggregated time series for other countries
    sns.lineplot(
        data=other_countries_data,
        x='Year',
        y='Value',
        hue='Country Name',
        palette='Set2',
        linewidth=1
    )

    # Customize title and labels
    plt.title(f"Comparison of India with Other Countries for {sdg} (UN SDG)", fontsize=10)
    plt.xlabel("Year", fontsize=6)
    plt.ylabel("Aggregated Value", fontsize=6)
    plt.legend(title='Country', loc='upper right')
    plt.xticks(rotation=45)

    # Display the plot
    plt.tight_layout()
    plt.show()
```

# **RANK CLASSIFICATION ON BASIS OF UN SDG**

```python
# Step 1: Aggregate values for each country and SDG category
agg_df = SDG.groupby(['Country Name', 'SDG'])['Value'].mean().reset_index()

# Step 2: Normalize the aggregated values (Min-Max normalization)
agg_df['Normalized Value'] = (agg_df['Value'] - agg_df['Value'].min()) / (agg_df['Value'].max() - agg_df['Value'].min())

# Step 3: Rank countries based on normalized values
agg_df['Rank'] = agg_df.groupby('SDG')['Normalized Value'].rank(ascending=False)  # Higher normalized value = better rank
```

```python
overall_scores = agg_df.groupby('Country Name')['Normalized Value'].mean().reset_index()
overall_scores['Overall Rank'] = overall_scores['Normalized Value'].rank(ascending=False)
overall_scores_sorted = overall_scores.sort_values(by='Overall Rank')
print(overall_scores_sorted)
```
